```python
dataset.insert(1,'theta0',1)

#Shuffle
dataset = dataset.sample(frac=20, replace=True, random_state=1)

x=dataset.iloc[:,[1,14,17,24,25]]
y=dataset.iloc[:,-1]
x1_min=x.iloc[:,1].min()
x1_max=x.iloc[:,1].max()

x2_min=x.iloc[:,2].min()
x2_max=x.iloc[:,2].max()

x3_min=x.iloc[:,3].min()
x3_max=x.iloc[:,3].max()

x4_min=x.iloc[:,4].min()
x4_max=x.iloc[:,4].max()


# normalizing data 

x.iloc[:,1]= (x.iloc[:,1]-x1_min)/(x1_max - x1_min)
x.iloc[:,2]= (x.iloc[:,2]-x2_min)/(x2_max - x2_min)
x.iloc[:,3]= (x.iloc[:,3]-x3_min)/(x3_max - x3_min)
x.iloc[:,4]= (x.iloc[:,4]-x4_min)/(x4_max - x4_min)

```

```python
# splitting data into train and test 

x_train=x[0:164]

y_train=y[0:164]

x_test=x[164:]

y_test=y[164:]

# transfer data from data frame into matrix

x_train=np.array(x_train)

y_train=np.array(y_train).flatten()


```

```python
# initially let theta1,2,3,4 , c = 0 

theta=np.array([0,0,0,0,0])

# m number of rows 

m=len(x_train)

# cost function 

def costFun(x_train,y_train,theta,m):
    cost = np.sum((x_train.dot(theta)-y_train)**2)/2/m
    return cost

l = 0.03                       # learning rate 
itr_num = 100                  # number of iterations 



def gradient_descent (x,y,theta,alpha,itr,m):
    
    #cost in every iteration
    cost_record = [0] * itr
    
    for i in range (itr):
        loss = (x.dot(theta))-y                  # (y actual - y predicted)
        gradient = x.T.dot(loss) / m             # 1/2m E (y actual - y predicted)
        theta = theta - (alpha *gradient )
        cost = costFun(x,y,theta,m)
        cost_record[i]= cost 
        
    return theta ,cost_record
    



(th,costR)=gradient_descent(x_train ,y_train,theta ,l,itr_num,m)


```

```python
#making prediction


y_predict = th[1] * x_test['curbweight'] + th[2] * x_test['enginesize'] + th[3] * x_test['citympg']+th[4] * x_test['highwaympg']+ th[0]

y_predict


```

```python
iteration = list(range (0,itr_num ))
x=np.array(costR)
y=np.array(iteration)
plot.plot(y,x)
```
